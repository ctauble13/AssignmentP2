```python
#compute cumulative distribution
####################
def survival_function_from_data (data):
    sorted_data = sorted(data)
    x = []
    cx = []
    norm = 0.0
    for n in sorted_data:
        ##this is for visualization purposes only
        x.append(n)
        cx.append(1.0 - norm/float(len(sorted_data)+1.0))
        norm = norm + 1.0
        x.append(n)
        cx.append(1.0 - norm/float(len(sorted_data)+1.0))
    return x, cx
#####################
```
