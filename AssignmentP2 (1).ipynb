```python
from scipy.stats import linregress
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

teams = ['crd','atl','rav','buf','car','chi','cin','cle','dal','den','det','gnb','htx',
         'clt','jax','kan','rai','sdg','ram','mia','min','nwe','nor','nyg','nyj','phi',
         'pit','sfo','sea','tam','oti','was']
turnovers=[]
wins=[]
avgOffensivePlays=[]


for team in teams:
    
    filein = 'Teams/' + str(team) + '.html'
    soup = BeautifulSoup(open(filein), 'lxml')

    entry = soup.find('tfoot', attrs={'class' : ''})
    winColumn=entry.find('td',attrs={'data-stat':'team_game_result'})
    winLoss=winColumn.get_text();
    win=winLoss[:winLoss.index("-")]
    
    tColumn = entry.find('td', attrs={'data-stat': 'turnovers'})
    pColumn= entry.find('td', attrs={'data-stat':"plays_offense"})
    turnover=tColumn.get_text();
    offenseivePlays=pColumn.get_text();
    avgPlays=round(int(offenseivePlays)/17)
    turnovers.append(turnover)
    wins.append(win)
    avgOffensivePlays.append(avgPlays)
    print(str(team)+": T/Os: "+turnover+" Wins "+win+"  Avg Offensive Plays " + str(avgPlays))
   
```

```python
from scipy.stats import linregress
data = [10, 20, 30, 40]
df = pd.DataFrame(data, columns=['Values'])

#Wins vs Turnovers
x1 = pd.DataFrame(turnovers,
y1 = wins
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)

# Second correlation: Wins vs Avg Offensive Plays
x2 = avgOffensivePlays
y2 = wins
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x2, y2)

# Create figure with two panels
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# ---- Plot 1: Wins vs Turnovers ----
axes[0].scatter(x1, y1)
axes[0].plot(x1, intercept1 + slope1*x1, 'r', label=f'y={slope1:.2f}x+{intercept1:.2f}')
axes[0].set_xlabel('Turnovers')
axes[0].set_ylabel('Wins')
axes[0].set_title(f'Wins vs Turnovers\nPearson r = {r_value1:.2f}')
axes[0].legend()

# ---- Plot 2: Wins vs Avg Offensive Plays ----
axes[1].scatter(x2, y2)
axes[1].plot(x2, intercept2 + slope2*x2, 'r', label=f'y={slope2:.2f}x+{intercept2:.2f}')
axes[1].set_xlabel('Avg Offensive Plays')
axes[1].set_ylabel('Wins')
axes[1].set_title(f'Wins vs Avg Offensive Plays\nPearson r = {r_value2:.2f}')
axes[1].legend()

plt.tight_layout()
plt.show()
```
